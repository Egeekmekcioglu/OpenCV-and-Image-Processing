```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("internal_external.png")
plt.imshow(img, cmap="gray"), img.shape
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
contours, hierarchy = cv2.findContours(img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE) 
external_contours = np.zeros(img.shape)
external_contours.shape
```

```python
len(contours), type(contours)
```

```python
type(hierarchy), hierarchy
```

```python
plt.imshow(img),
list(range(len(contours)))
```

```python
for i in range(len(contours)):
    
    
    # External Contours
    if hierarchy[0][i][3] == -1:
        
        cv2.drawContours(external_contours, contours, i ,255 ,-1)
        
```

```python
plt.imshow(external_contours, cmap="gray")
```

```python
internal_contours = np.zeros(img.shape)
internal_contours.shape

for i in range(len(contours)):
    
    
    # Internal Contours
    if hierarchy[0][i][3] != -1:
        
        cv2.drawContours(internal_contours, contours, i ,255 ,-1)
```

```python
plt.imshow(internal_contours, cmap="gray"), internal_contours.shape
```
